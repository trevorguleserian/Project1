```python
import pandas as pd
```

```python
tv_csv = '../Resources/tv_shows.csv.zip'
tv_df = pd.read_csv(tv_csv)
tv_df = tv_df[['Title', 'Year', 'IMDb', 'Netflix', 'Hulu', 'Prime Video', 'Disney+']]
tv_df
```

```python
clean_tv_df = tv_df.dropna(how='any').reset_index()
clean_tv_df
```

```python
clean_tv_df.to_csv('clean_tv.csv', index=False, header=True)
```
